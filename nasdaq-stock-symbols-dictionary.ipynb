```python
import os
import yfinance as yf
import pandas as pd
```

```python
# Path to your CSV files
csv_path = '/content/drive/MyDrive/final_project_deep_learning/nasdaq_dataset/data_nasdaq_csv/csv'

# Function to get list of CSV files
def get_csv_files(path):
    return [f for f in os.listdir(path) if f.endswith('.csv')]

# Function to get company info for a symbol (name and sector)
def get_company_info(symbol):
    try:
        ticker = yf.Ticker(symbol)
        company_name = ticker.info.get('longName', 'Unavailable')
        company_sector = ticker.info.get('sector', 'Unavailable')
        return company_name, company_sector
    except Exception as e:
        print(f"Error fetching data for {symbol}: {e}")
        return "Unavailable", "Unavailable"

# Get list of CSV files
csv_files = get_csv_files(csv_path)

# Extract symbols (remove .csv extension)
symbols = [os.path.splitext(file)[0] for file in csv_files]

# Create a DataFrame with symbols, company names, and sectors
company_data = []
for symbol in symbols:
    name, sector = get_company_info(symbol)
    company_data.append({'Symbol': symbol, 'Company Name': name, 'Sector': sector})

# Create a DataFrame
df = pd.DataFrame(company_data)

# Sort the DataFrame alphabetically by Company Name
df_sorted = df.sort_values(by='Symbol')

# Display the sorted results
print(df_sorted)

# Optionally, save to a CSV file in your Google Drive
output_path = '/content/drive/MyDrive/stock_symbols_names_sectors_sorted.csv'
df_sorted.to_csv(output_path, index=False)
print(f"Sorted results with sectors saved to '{output_path}'")
```

```python
import os
import pandas as pd
import yfinance as yf

# Path to your CSV files
csv_path = '/content/drive/MyDrive/final_project_deep_learning/nasdaq_dataset/data_nasdaq_csv/csv'

# Function to get list of CSV files
def get_csv_files(path):
    return [f for f in os.listdir(path) if f.endswith('.csv')]

# Function to get company info for a symbol (name, sector, and market cap, first trade date)
def get_company_info(symbol):
    try:
        ticker = yf.Ticker(symbol)
        company_name = ticker.info.get('longName', 'Unavailable')
        company_sector = ticker.info.get('sector', 'Unavailable')
        market_cap = ticker.info.get('marketCap', 'Unavailable')
        first_trade_date = ticker.info.get('firstTradeDateEpochUtc', 'Unavailable')
        return company_name, company_sector, market_cap, first_trade_date
    except Exception as e:
        print(f"Error fetching data for {symbol}: {e}")
        return "Unavailable", "Unavailable", "Unavailable", "Unavailable"

# Get list of CSV files
csv_files = get_csv_files(csv_path)

# Extract symbols (remove .csv extension)
symbols = [os.path.splitext(file)[0] for file in csv_files]

# Create a DataFrame with symbols, company names, sectors, and market cap
company_data = []
for symbol in symbols:
    name, sector, market_cap, first_trade_date = get_company_info(symbol)
    company_data.append({'Symbol': symbol, 'Company Name': name, 'Sector': sector, 'Market Cap': market_cap, 'First Trade Date': first_trade_date})

# Create a DataFrame
df = pd.DataFrame(company_data)

# Sort the DataFrame alphabetically by Symbol
df_sorted = df.sort_values(by='Symbol')

# Display the sorted results
print(df_sorted)

# Optionally, save to a CSV file in your Google Drive
output_path = '/content/drive/MyDrive/stock_symbols_dictionary.csv'
df_sorted.to_csv(output_path, index=False)
print(f"Sorted results with sectors and market cap saved to '{output_path}'")
```
